#### Generating random points within the Unit Square --- using the random library for the generation
#### Sorting the generated random points to retain uniformi

```python
import matplotlib.pyplot as plt
import random


#Generating the Unit Square ----
xUnitSquare = [0, 1, 0, 1]
yUnitSquare = [0, 1, 0, 1]
fig, ax = plt.subplots()
ax.fill(xUnitSquare, yUnitSquare)
plt.show()
```

```python
# Lets see how well the random.uniform works in the documentation.

count = 0
while(count < 10):
    x = random.uniform(0, 1)
    print(x)
    count+=1
```

```python
# ---- Good enough tbh -- Now lets Create the thing in 1a ---

def plotRandomPoints(N_SIZE):
    count = 0
    wX = []
    wY = []
    
    while count <= N_SIZE:
        # Limiting the size of the float returned by random
        wX.append(round(random.uniform(0,1),2))
        wY.append(round(random.uniform(0,1),2))
        count+=1
    # Sorting the returned lists so the points are uniformly distributed --- (;
    wX.sort()
    wY.sort()
    # Plotting the returned points
    plt.plot(wX, wY)
```

```python
# plotting n^3
plotRandomPoints(10000)
# showing the plot
plt.show()
```
